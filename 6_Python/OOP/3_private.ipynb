```python
# __init__  & magic methods __add__
# __a, ___fun() private var, func
# can be accessed using 
# _<class_name>__a, _<class_name>__fun()
```

```python
class customer:
    def __init__(self,name) -> None:
        self.name=name

def change(cust):
    cust.name='mohit'


cust1=customer("nitin")
print(cust1.name)
change(cust=cust1)
print(cust1.name)

```
